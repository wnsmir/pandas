#20. 데이터를 로드하라

```python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv'
df = pd.read_csv(DataUrl)
df.head()
```

#21. quantity의 값이 3인 데이터를 추출하여 첫 5행을 출력하라

```python
df[df.quantity == 3].head()# 이 코드의 뜻은 df의 열중 quantity를 뽑아내 그 값이 3인 true값의 index만 뽑아낸다
# 각 컬럼은 df의 메서드로 들어가는데 좋은 방법은 아님. 컬럼과 메서드가 겹칠 수 있기 때문
```

```python
df.loc[df['quantity'] == 3].head()#loc를 쓰는 습관을 들이자.
```

#22. quantity컬럼값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행을 출력하라

```python
df.loc[df['quantity']==3].reset_index(drop=True)
```

#23. quantity, item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하라

```python
df_new = df.loc[:,['quantity', 'item_price']].copy()#.copy를 쓰는 습관을 들이자
df_new
```

#24. item_price 컬럼의 달러표시 문자를 제거하고 float타입으로 저장하여 new_price 컬럼에 저장하라

```python
df.loc[:,'new_price'] = df['item_price'].str.replace('$','').astype(float)#str이 있어야 특정 문자의 일부요소만 대체가능
df
```

#25. new_price 컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수를 구하여라

```python
ans = len(df.loc[df['new_price'] <= 5])
ans
```
